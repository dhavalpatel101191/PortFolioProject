```python
import pandas as pd
import numpy as np
import datetime
from pandas.tseries.offsets import MonthEnd
```

```python
Covid = pd.read_csv('./owid-covid-data.csv')
```

```python
Covid.head()
```

```python
Covid.columns
```

```python
Covid.columns

Column = ['continent', 'location', 'date', 'total_cases', 'new_cases','total_deaths', 'new_deaths','icu_patients','hosp_patients','people_fully_vaccinated',
       'people_vaccinated']
```

```python
Date_Locaton_Rate = pd.DataFrame()
```

```python
Date_Locaton_Rate['Positive Rate']= Covid['positive_rate']
```

```python
Date_Locaton_Rate['Location']= Covid['location']
Date_Locaton_Rate['Date']= Covid['date']
Date_Locaton_Rate['Continent'] = Covid['continent']
```

```python
Date_Locaton_Rate.head()
```

```python
Date_Locaton_Rate.dropna(subset=['Continent'])
```

```python
Date_Locaton_Rate.to_excel('Date_Locaton_Rate.xlsx',index=False)
```

```python
Covid_New = pd.DataFrame()
```

```python
Covid_New[['Continent','Location',
'Date','Total_cases','New_cases','Total_deaths',
 'New_deaths','ICU_patients','Hosp_patients','People_fully_vaccinated','People_vaccinated']] = Covid[Column]
```

```python
Covid_New
```

```python
Covid_New['Date'] = pd.to_datetime(Covid_New['Date'])
```

```python
Covid_Copy = Covid_New
```

```python
Covid_Copy
```

```python
Continent_Data = Covid_Copy.dropna(subset=['Continent'])
```

```python
Continent_Data
```

```python
Population = pd.read_excel('./Table_3.xlsx')
```

```python
Population.head()
```

```python
Covid_Data_Final = pd.merge(left=Continent_Data,right=Population,how='left',on='Location')
```

```python
Covid_Data_Final.to_excel('Continent_Data.xlsx',index=False)
```

```python
Covid_Data_Final
```

```python
People_Fully_Vaccinated = pd.DataFrame()
```

```python
People_Fully_Vaccinated = Covid_Data_Final.groupby(['Location'])[['People_fully_vaccinated','Total_cases','Total_deaths']].max()
```

```python
People_Fully_Vaccinated
```

```python
People_Fully_Vaccinated.to_excel('People_Fully_Vaccinated.xlsx')
```
