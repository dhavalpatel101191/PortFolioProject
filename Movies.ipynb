```python
# Import Libraries
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Data
Movie = pd.read_csv('./Data/movies.csv')
```

```python
# Quick look of Data 
Movie.head()
```

```python
Movie_Copy = Movie
```

```python
Movie.dtypes

```

```python
# Drop Null values
Movie_Copy.dropna(inplace=True)
```

```python
# Change Data types of Columns
Movie_Copy['budget'] = Movie_Copy['budget'].astype('int64')
Movie_Copy['gross'] = Movie_Copy['gross'].astype('int64')
Movie_Copy['votes'] = Movie_Copy['votes'].astype('int64')
Movie_Copy['runtime'] = Movie_Copy['runtime'].astype('int64')  
```

```python
Movie_Copy.head()
```

```python
# Get Correct year from Released Column
Actual_Year = Movie_Copy['released'].str.split(',',expand=True)
```

```python
# Create Corret year column
Movie_Copy['Year_Corrected'] = Actual_Year[1].str.strip().str[:4]
```

```python
Movie_Copy.head()
```

```python
# Sort values by Gross income of Movie
Movie_Copy.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# Drop duplicayes values
Movie_Copy.drop_duplicates()
```

```python
# Budget High correlation
# Company high Correlation

# Scatter Plot with Budget vs Gross

plt.scatter(x=Movie_Copy['budget'],y=Movie_Copy['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earning')
plt.show
```

```python
# Regression plot

sns.regplot(x='budget',y='gross',data=Movie_Copy,scatter_kws={"color":"green"},line_kws={"color":"blue"})
```

```python
Movie_Copy.corr()
```

```python
Corr_Mat = Movie_Copy.corr()

sns.heatmap(Corr_Mat,annot=True)
plt.show()
```
