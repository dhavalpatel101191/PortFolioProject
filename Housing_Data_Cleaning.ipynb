```python
import pandas as pd
```

```python
House = pd.read_excel('Nashville Housing Data for Data Cleaning.xlsx')
```

```python
House.head()
```

```python
House.info()
```

/-------------Fill Null Value In PropertyAddress Column by Forward Fill Method------------------/

```python
House['PropertyAddress'].fillna(method='ffill',inplace=True)
```

```python
House.info()
```

/------------------------------Break PropertAddress Column in Two Different Column -----------------/

```python
PropertySplit = House['PropertyAddress'].str.split(',',expand=True)
```

```python
PropertySplit
```

```python
House[['SplitPropertyAddress','SplitPropertyCity']] = PropertySplit
```

```python
House.head()
```

/------------------------------Break OwnerAddress Column in Three Different Column -----------------/

```python
OwnerSplit = House['OwnerAddress'].str.split(',',expand=True)
```

```python
OwnerSplit.head()
```

```python
House[['SplitOwnerAddress','SplitOwnerCity','SplitOwnerState']] = OwnerSplit
```

```python
House
```

/------------------------------Change Values in particular Column ------------------------------/

```python
House['SoldAsVacant'].unique()
```

```python
House_Copy = House
```

```python
House_Copy['SoldAsVacant'].replace('N','No',inplace=True)
```

```python
House_Copy['SoldAsVacant'].replace('Y','Yes',inplace=True)
```

```python
House_Copy['SoldAsVacant'].unique()
```

/-------------------------Drop Duplicates in Data with particular Column-----------------------/

```python
House_Copy.drop_duplicates(subset=['ParcelID',
             'PropertyAddress',
			 'SalePrice',
			 'SaleDate',
			 'LegalReference'],inplace=True)
```

```python
House_Copy
```

/-----------------------Drop Unnecessory Columns---------------------/

```python
House_Copy.drop(columns=['OwnerAddress','TaxDistrict','PropertyAddress'])
```

/----------------------------------------Rename Columns Name-------------------------------------------/

```python
House_Copy.rename(columns={
    'SplitPropertyAddress': 'PropertyAddress',
    'SplitPropertyCity' : 'PropertyCity',
    'SplitOwnerAddress' : 'OwnerAddress',
    'SplitOwnerCity' : 'OwnerCity',	
    'SplitOwnerState' : 'OwnerState'
})
```

/------------------------------Create Final CSV file from Data set--------------------------------------/

```python
House_Copy.to_csv('Cleaned_Nashville_Housing_Data.csv',index=False)
```
